```python
import os
import pandas as pd
import re
```

```python
def get_txt_files(directory):
    txt_files = []
    for file in os.listdir(directory):
        if file.endswith(".gff3"):
            txt_files.append(file)
    return txt_files

file_list = get_txt_files("/data1/NFS/home/whx/P450_database/bitacore/result/gff/")

prot_dict = {}

for file in file_list:
    file_name = "/data1/NFS/home/whx/P450_database/bitacore/result/gff/" + file
    tax_name = file.split("_")[0] + "_" + file.split("_")[1]
    prot_dict[tax_name] = []
    with open(file_name,"r") as lines:
        for line in lines:


            items = line.split("\t")

            if items[2] == "mRNA" and items[1] == "AnnotGFF":
                match = re.search(r'ID=(.*?);', items[8]).group(1)
                if re.search(r'_split\d+', match):
                    match = match.rstrip("split0123456789").rstrip("_")
                if match not in prot_dict[tax_name]:
                    prot_dict[tax_name].append(match)
                

def merge_fa_files(folder_path, output_file):
    with open(output_file, 'w') as outfile:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                if file.endswith('.faa'):
                    filepath = os.path.join(root, file)
                    with open(filepath) as infile:
                        outfile.write(infile.read())

for k,j in prot_dict.items():
    with open("/data1/NFS/home/whx/filt_seq/" + k + "_filt.fa","w") as fa_file:
        for seq_name in prot_dict[k]:
            if len(all_prot_dict[seq_name]) > 100 and len(all_prot_dict[seq_name]) <1000:
                fa_file.write(">" + seq_name + '\n')
                fa_file.write(all_prot_dict[seq_name] + '\n')


merge_fa_files("/data1/NFS/home/whx/P450_database/bitacore/result/protein", "/data1/NFS/home/whx/P450_database/bitacore/result/protein/all_pep.fa")

def parse_fa_file(file_path):
    fa_dict = {}
    
    with open(file_path, 'r') as file:
        current_header = None
        current_sequence = ''
        
        for line in file:
            line = line.strip()
            
            if line.startswith('>'):

                if current_header is not None:
                    # 将上一个序列的信息添加到字典中
                    fa_dict[current_header] = current_sequence
                

                current_header = re.search(r'\wP_\d+\.\d+', line).group()
                current_sequence = ''
            else:

                current_sequence += line
        

        if current_header is not None and len(current_sequence) > 0:
            fa_dict[current_header] = current_sequence
    
    return fa_dict


all_prot_dict = parse_fa_file("/data1/NFS/home/whx/P450_database/bitacore/result/protein/all_pep.fa")





```
